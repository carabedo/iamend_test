```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
zv=np.linspace(-1.5,0,10)*1e-3
```

```python
simbols=['-k','-r','-b','-g']
```

```python
zv=np.linspace(-1.5,0,20)*1e-3
plt.figure(figsize=[10,6])
for m,muestra in enumerate(names):
    jv=[]
    for z in zv:
        jv.append(np.abs(ci.theo.jhf(0.0018,z,1,f2,ci.bo.m1,sigmas[m],mues2[m])))
    plt.plot(np.array(jv)/np.array(jv).max(),zv/1e-3,simbols[m])
    plt.title('J normalizada 95 kHz')
    plt.xlabel('J/Jmax')
    plt.ylabel('z [mm]')
    plt.legend(names)
    plt.grid(True)
```

```python
simbols=['-k','-r','-b','-g']
simbolst=['--k','--r','--b','--g']
```

```python
labels=['M1010', 'M1010','M1022','M1022', 'M347','M347', 'M1566','M1566']

```

```python
zv=np.linspace(-3,0,20)*1e-3
fig=plt.figure(figsize=[10,8])
for m,muestra in enumerate(names):
    jv=[]
    jteov=[]
    for z in zv:
        jv.append(np.abs(ci.theo.jhf(0.0018,z,1,f1,ci.bo.m1,sigmas[m],mues1[m])))
        jteov.append(np.abs(jteo(z,f1,mues1[m],sigmas[m])))
    plt.plot(np.array(jv)/np.array(jv).max(),zv/1e-3,simbols[m])
    plt.plot(jteov,zv/1e-3,simbolst[m],label='_nolegend_')
plt.title('J normalizada 15 kHz',fontsize=25)
plt.xlabel('J/Jmax',fontsize=20)
plt.ylabel('z [mm]',fontsize=20)
plt.legend(names,loc='center',fontsize=15)
ax = fig.gca()
ax.xaxis.tick_top()
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
```

```python
zv=np.linspace(-1,0,20)*1e-3
fig=plt.figure(figsize=[10,8])
for m,muestra in enumerate(names):
    jv=[]
    jteov=[]
    for z in zv:
        jv.append(np.abs(ci.theo.jhf(0.0018,z,1,f2,ci.bo.m1,sigmas[m],mues2[m])))
        jteov.append(np.abs(jteo(z,f2,mues2[m],sigmas[m])))
    plt.plot(np.array(jv)/np.array(jv).max(),zv/1e-3,simbols[m])
    plt.plot(jteov,zv/1e-3,simbolst[m],label='_nolegend_')
    
plt.title('J normalizada 95 kHz',fontsize=25)
plt.xlabel('J/Jmax',fontsize=25)
plt.ylabel('z [mm]',fontsize=25)
plt.legend(names,fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)

```

```python
def jteo(z,f,mur,sigma):
    k=(1+1j)/np.sqrt(2/(2*np.pi*f*mur*mu0*sigma))
    return np.exp(k*z)
```

```python
jvteo=[]
for n,z in enumerate(zv):
    jvteo.append(np.abs(jteo(z,f1,mues1[0],sigmas[0])))
```
